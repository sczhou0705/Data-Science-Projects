```python
# Check every column and verify if the dataset have duplicated columns
def finding_duplicate_columns(data):
    col_name_list = [x for x in data.columns]
    print("length: ",len(col_name_list))
    duplicate_col2 = []
    i = 0
    while i<=len(col_name_list)-1:
        
        for j in range (1,len(col_name_list)):
            if data[col_name_list[i]].equals(data[col_name_list[j]]) == False:
                continue
            else:
                duplicate_col2.append((col_name_list[i],col_name_list[j]))
        col_name_list.pop(0)
    return duplicate_col2
duplicate_col = finding_duplicate_columns(df)
duplicate_col
   
```

```python
# Scale the numercial data
```

#### Categorical data

```python
df_cat.shape
```

```python
df_cat
```

```python
# Check how many labels in each categorical columns
for col in df_cat.columns:
    print(col,":",len(df_cat[col].unique()),' labels')
```

```python
# Take city as example to check the sorting.
df_cat.city.value_counts().sort_values(ascending=False).head()
```

The city attribute has 64 labels from the above output, and I only filter the city code through the top five frequencies and encode them.

```python
# Create function for storing the label as list
def label_list(col):
    col_list = [x for x in col.value_counts().sort_values(ascending=False).head().index]
    return col_list

type_A_list = label_list(df_cat.type_A)
type_B_list = label_list(df_cat.type_B)
city_list = label_list(df_cat.city)
loyalty_list = label_list(df_cat.loyalty)
print(type_A_list)
print('---')
print(type_B_list)
print('---')
print(city_list)
print('---')
print(loyalty_list)
```

```python
# Encoding Type A
for label in type_A_list:
    df_cat[label] = np.where(df_cat['type_A']==label,1,0)
df_type_A = df_cat[['type_A']+type_A_list]
df_type_A = df_type_A.drop(df_type_A.columns[0],axis=1)
df_type_A.columns = ['type_A_3','type_A_0','type_A_6']
df_type_A.head(1)
```

```python
# Encoding Type B
for label in type_B_list:
    df_cat[label] = np.where(df_cat['type_B']==label,1,0)
df_type_B = df_cat[['type_B']+type_B_list]
df_type_B = df_type_B.drop(df_type_B.columns[0],axis=1)
df_type_B.columns = ['type_B_3','type_B_0','type_B_6','type_B_9']
df_type_B.head(1)
```

```python
# Encoding city
for label in city_list:
    df_cat[label] = np.where(df_cat['city']==label,1,0)
df_city = df_cat[['city']+city_list]
df_city = df_city.drop(df_city.columns[0],axis=1)
df_city.columns = ['city_2','city_0','city_8','city_1','city_9']
df_city.head(1)
```

```python
# Encoding loyalty
for label in loyalty_list:
    df_cat[label] = np.where(df_cat['loyalty']==label,1,0)
df_loyalty = df_cat[['loyalty']+loyalty_list]
df_loyalty = df_loyalty.drop(df_loyalty.columns[0],axis=1)
df_loyalty.columns = ['loyalty_99','loyalty_3','loyalty_1','loyalty_2','loyalty_0']
df_loyalty.head(1)
```

```python
# Update df_cat 
df_cat_update = pd.concat([df_type_A,df_type_B,df_city,df_loyalty,df_cat[['lor_M']]],axis=1)
df_cat_update
```

```python
# Select features for categorical data
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = df_cat_update
y = df_res
print("Feature data dimension: ", X.shape)
print("----")
select = SelectKBest(score_func=chi2, k=15)
fit = select.fit(X,y)
# type(fit)
scores = pd.DataFrame(fit.scores_)
columns = pd.DataFrame(X.columns)
# Concat two dataframes for better visualization 
featureScores = pd.concat([columns,scores],axis=1)
featureScores.columns = ['Features','Score'] 
# Print 8 best features
print(featureScores.nlargest(15,'Score'))
```

- Even if loyalty_99 gets a 311.27 score, we would not consider it because 99 means unclassified values.
- We filter out the top 15 scores. Some are relevant to product type, some are related to the relationship between clients and the insurance company, and some are city code.

## Construct model

```python
# Creating train and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
y_train.values.sum()
```

```python
# Get the percentage of the training set of bought product
a = y_train.values.sum() / y_train.shape[0]
a
```

```python
null_error_rate = (1-a)
null_error_rate * 100
```

Since we know that 27.55% of the clients in the training set bought the product, we could achieve a training model accuracy of ( 1-0.27549) = 72.33% by simply predicting " did not buy" for each client. It is a null error rate. If the following accuracy of a binary logistic regression model does not exceed the invalid error rate, the model is none of any value.

#### Model 1

```python
y_train.values.sum()
```

```python
# Get the percentage of the training set of bought product
y_train.values.sum() / y_train.shape[0]
```

```python
X1 = df_model1
y = df_res
```

```python
# Define a vector containing the names of the attributes to use
predictors = [x for x in df_model1.columns]
# Create a subset of the training data using ONLY the selected explanatory variables
X1_train = X1[predictors].values
# Create a subset of the testing data using ONLY the selected explanatory variables
X1_test = X1[predictors].values
# Isolate the response indicator for the training data
y_train = y.values

# Sanity check on training data
X1_train[:5]
```

```python
# Sanity check on response indicator
y_train[:5]
```

```python
# We're using the LogisticRegression() method for this model
model = LogisticRegression()

# Fit the model: X_train contains our explanatory variables while 
# y_train contains the response variable
model.fit(X1_train, y_train)
# Calculate the accuracy of the model relative to the training data set
model.score(X1_train, y_train)
```

The model we've generated has an accuracy score of 76.44%, over the null error rate of 72.33%. The model appears to be useful.

```python
# Examine the model coefficients for the explanatory variables
print(predictors)
model.coef_
```

From the above we see that:
- #### turnover_B: 
An increase in the value of 'turnover_B' is associated with an decreased likelihood of buying the product from the insurance company.
- #### turnover_A: 
An increase in the value of 'turnover_A' is associated with an increased likelihood of buying the product from the insurance company.
- #### age_P: 
The order the clients, the more likely they bought.
- #### lor_m: 
The more time relationship between clients and insurance company, the more likely they bought. 
- #### type_B_0: 
An increase in the value of 'type_B_0' is associated with an increased likelihood of buying the product from the insurance company.
- #### type_A_0:
An increase in the value of 'type_A_0' is associated with an increased likelihood of buying the product from the insurance company.
- #### loyalty_3:
An increase in the value of 'loyalty_3' is associated with a ndecreased likelihood of buying the product from the insurance company.


```python
# Cross validation.
accuracy = cross_val_score(model, X1_train, y_train, scoring = "accuracy", cv = 10)
recall = cross_val_score(model, X1_train, y_train, scoring = "recall", cv = 10)
auc = cross_val_score(model, X1_train, y_train, scoring = "roc_auc", cv = 10)
f1 = cross_val_score(model, X1_train, y_train, scoring = "f1", cv = 10)
precision = cross_val_score(model, X1_train, y_train, scoring = "precision", cv = 10)
print(accuracy.mean())
print('---')
print(recall.mean())
print('---')
print(auc.mean())
print('---')
print(f1.mean())
print('---')
print(precision.mean())
print('---')
model1_summary = [accuracy.mean(),recall.mean(),auc.mean(),f1.mean(),precision.mean()]
model1_summary = pd.DataFrame(model1_summary,columns=['Model1'],index=['accuracy','recall','auc','f1','precision'])
model1_summary
```
